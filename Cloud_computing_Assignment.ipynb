```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data from 'books.csv' into a DataFrame
book_data = pd.read_csv('books.csv')

# Select specific columns of interest for our analysis
columns_of_interest = ['original_title', 'authors', 'average_rating', 'ratings_count', 
                       'ratings_1', 'ratings_2', 'ratings_3', 'ratings_4', 'ratings_5']
hp_data = book_data[columns_of_interest]

# Filter for J.K. Rowling's Harry Potter books
hp_books = hp_data[hp_data['authors'].str.contains('J.K. Rowling') & 
                   hp_data['original_title'].str.contains('Harry Potter')]
```

```python
# Plotting the distribution of ratings for each book
plt.figure(figsize=(12, 8))
for index, row in hp_books.iterrows():
    ratings = [row[f'ratings_{i}'] for i in range(1, 6)]
    plt.plot(ratings, marker='s', label=row['original_title'])

plt.xlabel('Rating Stars')
plt.ylabel('Number of Ratings')
plt.title('Distribution of Ratings for Harry Potter Books')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Plotting the total ratings count for each book to find the best selling one
sns.set_theme(style="darkgrid")
plt.figure(figsize=(12, 8))
sns.barplot(data=hp_books, x='original_title', y='ratings_count', palette='coolwarm')
plt.xlabel('Book Title')
plt.ylabel('Total Ratings Count')
plt.title('Total Ratings Count for Harry Potter Books')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Plotting the average rating for each book with a line plot using weighted average for the series
plt.figure(figsize=(12, 8))
sns.lineplot(data=hp_books, x='original_title', y='average_rating', marker='o', color='green')
plt.axhline(hp_books['average_rating'].mean(), ls='--', color='red', label='Series Average Rating')
plt.xlabel('Book Title')
plt.ylabel('Average Rating')
plt.title('Average Rating for Harry Potter Books')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Plotting the proportion of 5-star ratings for each book to see which book is liked the most
hp_books['proportion_5_stars'] = hp_books['ratings_5'] / hp_books['ratings_count']
plt.figure(figsize=(12, 8))
sns.barplot(data=hp_books, x='original_title', y='proportion_5_stars', palette='pastel')
plt.xlabel('Book Title')
plt.ylabel('Proportion of 5-Star Ratings')
plt.title('Proportion of 5-Star Ratings for Harry Potter Books')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
